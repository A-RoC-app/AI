```python
# CIFAR-10 데이터셋 클래스 이름 리스트
cifar10_class_names = [
    "kickboard", "sidewalk block"
]

# 레이블 파일 작성
label_file_path = "cifar10_labels.txt"

with open(label_file_path, "w") as f:
    for i, class_name in enumerate(cifar10_class_names):
        f.write(f"{i} {class_name}\n")

print(f"레이블 파일이 {label_file_path}에 생성되었습니다.")
```
